#三维投影例子

```python
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
for c, z in zip(['r', 'g', 'b', 'y'], [30,20,10,0]):
    x=np.arange(20)
    y=np.random.rand(20)
    cs = [c] * len(x)
    ax.bar(x,y,z,zdir='y',color=cs)

plt.show()
```

#二维均匀分布
$f(x,y)$=$1/A$,$(x,y)$ in $G$

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(0,1,0.01)
Y = np.arange(0,1,0.01)
X,Y=np.meshgrid(X,Y)
Z1=1
Z2=0
surf=ax.plot_surface(X,Y,Z1,color='b')
surf=ax.plot_surface(X,Y,Z2,color='r')

plt.show()
```

#二维正态分布:bivariate_normal

```python
from pylab import *
x = np.linspace(-10,10,200)
y=x
X,Y = meshgrid(x,y)
Z=bivariate_normal(X,Y,1,5,0,-5,0)
fig = figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='OrRd')
```

#二维非经典分布

$f(x,y)$=$6e^{-(2x+3y)}$, $x>0, y>0$

```python
x_arr = np.linspace(0.001,1,200)
y_arr = x_arr
X,Y = meshgrid(x_arr,y_arr)
XY=zip(X,Y)

def get_z(x, y):
    return 6*(np.e**-(2*x + 3*y))

def get_z_meshgrid(x_corr_list, y_corr_list):
        z_row_meshgrid = []
        
        for i, x_corr in enumerate(x_corr_list):
            z_row_meshgrid.append(get_z(x_corr, y_corr_list[i]))
        
        return z_row_meshgrid

Z=map(lambda xy: get_z_meshgrid(*xy), XY)

fig = figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='OrRd')
```
